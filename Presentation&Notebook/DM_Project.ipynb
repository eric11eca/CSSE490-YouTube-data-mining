```python
import pandas as pd
from datetime import datetime
```

```python
df= pd.read_csv('./INvideos.csv')
```

```python
df.head()
```

```python
# df['trending_date'] = 
df['trending_date'] = df['trending_date'].apply(lambda x: datetime.strptime(x, '%y.%d.%m'))
df.head()
```

```python
df['publish_time'] = df['publish_time'].apply(lambda x: datetime.strptime(x, '%Y-%m-%dT%H:%M:%S.000Z'))
```

```python
df.head()
# df['trending_date'] = pd.to_datetime(df['trending_date']) 
```

```python
top_channel = df.groupby('channel_title')['likes'].sum()
top_channel.sort_values(ascending=False).head(10).plot.bar(title='top 10 most liked channels')
```

```python
top_channel = df.groupby('channel_title')['comment_count'].sum()
top_channel.sort_values(ascending=False).head(10).plot.bar(title='top 10 most commented channels')
```

```python
top_channel = df.groupby('channel_title')['views'].sum()
top_channel.sort_values(ascending=False).head(10).plot.bar(title='top 10 most viewed channels')
```

```python
volume=pd.pivot_table(df,values='views',index =['trending_date'],aggfunc='sum')
ax1 = volume.plot.line(color='r')
volume.rolling(10).mean().plot.line(color='b',ax=ax1,title='trending for total views of each date' )
```

```python
import json
f = open("./IN_category_id.json")
data = f.read()
f.close()
out = json.loads(data)
categories = pd.DataFrame([
    {'category_id':x['id'],
     'category':x['snippet']['title'],
     'assignable':x['snippet']['assignable']
    } for x in out['items']
])
categories['category_id'] = categories['category_id'].astype(int)
categories.head()
```

```python
categories.shape
```

```python
df['category_id'] = df['category_id'].astype(int)
df = df.merge(categories, on='category_id')
df.drop(['assignable','category_id'], axis=1)
df.head()
```

```python
df.groupby('category').likes.mean().plot(kind="bar", figsize=(10,5))
```
